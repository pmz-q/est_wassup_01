```python
import pandas as pd
```

```python
ROOT_PATH = '../../data/'
ROOT_DIR = '01-1/'
ROOT_DIR_ORIGIN = '01/'

TRAIN_FILE_LIST = ['2019.xls', '2020.xls', '2021.xls']
TEST_FILE = '2022.xls'
```

```python
df_2022 = pd.read_html(f'{ROOT_PATH}{ROOT_DIR}{TEST_FILE}')[0]
```

```python
df_2022 = df_2022.reset_index()
```

```python
df_2019 = pd.read_html(f'{ROOT_PATH}{ROOT_DIR}{TRAIN_FILE_LIST[0]}')[0]
df_2020 = pd.read_html(f'{ROOT_PATH}{ROOT_DIR}{TRAIN_FILE_LIST[1]}')[0]
df_2021 = pd.read_html(f'{ROOT_PATH}{ROOT_DIR}{TRAIN_FILE_LIST[2]}')[0]
```

```python
df_train = pd.concat([df_2019, df_2020, df_2021]).reset_index()
df_train.head()
```

## 수집한 데이터와 기존 데이터가 같은 데이터임을 확인합니다.
- 형태가 다른 데이터는 제외합니다.

### Train data

```python
# test if same data from origin
train_origin = pd.read_csv(f'{ROOT_PATH}{ROOT_DIR_ORIGIN}/train.csv')
train_origin.head()
```

```python
check_cols = ['요일', '기상상태', '시군구', '도로형태', '노면상태', '피해운전자 연령', '법규위반', '사망자수', '피해운전자 연령']
train_origin.loc[:, check_cols].equals(df_train.loc[:, check_cols])
```

### Test data
- 사용이 어렵다.
- 기존의 2022 데이터에서 몇 데이터들을 제외한 것 같다.

```python
test_origin = pd.read_csv(f'{ROOT_PATH}{ROOT_DIR_ORIGIN}/test.csv')
test_origin.head()
```

```python
check_cols = ['요일', '기상상태', '시군구', '도로형태', '노면상태']
test_origin.loc[:10, check_cols].equals(df_2022.loc[:10, check_cols])
```

```python
test_origin.loc[:, check_cols].equals(df_2022.loc[:, check_cols])
```

## 결론
- train 데이터만 csv 로 저장, 및 추후 시간이 남으면 사건ID 를 기준으로 추가 데이터를 찾아본다.
